# Leitura da Base

```python
#Código básico para a abertura e leitura do arquivo trabalhado
df = pd.read_csv("./ifood-restaurants-november-2020.csv", sep =",", header = 0)
```

# Explorando as Variáveis

```python
#df.dtypes = serve apenas para identificar os tipos (classificação) das variáveis
df.dtypes
```

```python
#"df.info()" serve para trazer um overview da tabela, informando a quantidade de dados dispníveis (rangeindex), mas também mostrando os dados nulos por colunas.
df.info()
```

```python
#"df.select_dtypes(include=["number"])": teve objetivo de mostrar as colunas que possuem caracteres numéricos, para que fosse possível a classificação em variável numérica discreta ou contínua. Apenas a variável delivey_fee é discreta, já que é composta por números inteiros.
df.select_dtypes(include=["number"])
```

```python
#De forma similar, é utilizado o código abaixo para mostrar as colunas que possuem dados do tipo texto. Em se tratando de pandas, podem ser classificados como "object" ou "string". Após isso, é possível classificar as variáveis categóricas entre nominais e ordinárias. No caso, apenas o price_range é classificado como ordinária.
df.select_dtypes(include = ["object", "string"])
```

```python
#Print das duas catagorias de variáveis (numéricas e categóricas)

numericas = df.select_dtypes(include = ["number"]).columns.tolist()
categoricas = df.select_dtypes(include = ["string", "object"]).columns.tolist()

print("Variáveis numéricas:", numericas)
print("Variáveis categóricas: ", categoricas)
```
